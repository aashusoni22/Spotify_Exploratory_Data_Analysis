# Spotify - Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import seaborn as sns
```

```python
# Importing sotify tracks dataset

df_tracks = pd.read_csv('C:/Users/omson/Downloads/Spotify Dataset/tracks.csv')

df_tracks.head()
```

```python
# checking null values

pd.isnull(df_tracks).sum()
```

```python
# fetching detail info

df_tracks.info()
```

```python
# looking at descriptive statistics

df_tracks.describe().transpose()
```

## 10 least & most popular songs on Spotify

```python
# fidning 10 least popular songs on spotify

least_pop = df_tracks.sort_values('popularity', ascending = True).head(10)

least_pop
```

```python
# finding 10 most popular songs on spotify

most_pop = df_tracks.query('popularity > 90', inplace = False).sort_values('popularity',ascending = False)

most_pop[:10]
```

## Setting Index

```python
# setting release_date as index(permanently) for better analysis

df_tracks.set_index('release_date', inplace = True)

# converting index columns's data type to datetime format

df_tracks.index = pd.to_datetime(df_tracks.index)

df_tracks.head()
```

```python
# looking at artist present at 18th row

df_tracks[['artists']].iloc[18]
```

## Converting duration from miliseconds to seconds

```python
# Now converting duration from milisecond to seconds

df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
df_tracks.duration.head()
```

## Visualizing data 

```python
# Finding Correlation between variables using Visualization

corr_var = df_tracks.drop(['key','mode','explicit'], axis = 1).corr(method = 'pearson')
plt.figure(figsize = (14, 6))
heatmap = sns.heatmap(corr_var, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='black')
heatmap.set_title("Correlation Heatmap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
# Using only 0.4 percent from total data by creating a sample

sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
# Now let's create a regression plot for Loudness vs Energy to find Correlation

plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', color = 'c').set_title("Loudness vs Energy Correlation")
```

```python
# Creating a regression plot for Popularity vs Acousticness to find Correlation

plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', color = 'g').set_title("Popularity vs Acousticness Correlation")
```

```python
# Creating new columns "years" from release_date column

df_tracks['year'] = df_tracks.index.get_level_values('release_date')
df_tracks.year = pd.to_datetime(df_tracks.year)
years = df_tracks.year.dt.year
```

```python
# Now Creating distribution plot (histogram) to visualize number of songs per year

sns.displot(years, discrete=True,aspect=2,height=6,kind='hist', color='olivedrab').set(title="Number of songs per year on spotify")
```

```python
# Creating bar chart to visualize total duration per year

total_dur = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dur, ax = ax, errwidth = False, color='limegreen').set(title = 'Year vs Duration')
plt.xticks(rotation = 90)
```

```python
# Creating Line chart to visualize total duration per year

total_dur = df_tracks.duration
sns.set_style(style = 'whitegrid')
fig_dims = (14, 8)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dur, ax = ax).set(title = 'Year vs Duration')
plt.xticks(rotation = 60)
```

```python
# Importing .csv file

df_genre = pd.read_csv('C:/Users/omson/Downloads/Spotify Dataset/SpotifyFeatures.csv')
```

```python
# Comparing durations of all genres using barplot

plt.title("Duration of the Songs per Genre")
sns.barplot(data=df_genre, y='genre', x='duration_ms', palette='inferno_r')
plt.xlabel('Duration in miliseconds')
plt.ylabel('Genres')
```
